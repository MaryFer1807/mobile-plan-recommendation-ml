```python
# Importamos las librer√≠as necesarias
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
# Cargar el dataset
df = pd.read_csv('/datasets/users_behavior.csv')

# Mostrar las primeras filas
df.head()
```

```python
# Informaci√≥n general sobre los datos
df.info()

# Estad√≠sticas b√°sicas
df.describe()
```

```python
# Caracter√≠sticas (features)
X = df.drop('is_ultra', axis=1)

# Objetivo (target)
y = df['is_ultra']
```

```python
# Primero, dividir en entrenamiento+validaci√≥n y prueba
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2, random_state=12345)

# Ahora dividir entrenamiento+validaci√≥n en entrenamiento y validaci√≥n
X_train, X_valid, y_train, y_valid = train_test_split(X_temp, y_temp, test_size=0.25, random_state=12345)
# 0.25 x 0.8 = 0.2, as√≠ que: 60% train, 20% valid, 20% test
```

```python
print(X_train.shape, X_valid.shape, X_test.shape)
```

<div class="alert alert-block alert-success">
<b>Comentario del revisor (1ra Iteraci√≥n)</b> <a class=‚ÄútocSkip‚Äù></a>

Hiciste una correcta segmentaci√≥n del dataset en los subconjuntos de entrenamiento, validaci√≥n y prueba, lo cual permite evaluar el rendimiento real del modelo y prevenir el sobreajuste. Esta separaci√≥n garantiza que el modelo generalice bien y no solo funcione con los datos que utiliz√≥ para aprender.

</div>

<div class="alert alert-block alert-warning">
<b>Comentario del revisor (1ra Iteraci√≥n)</b> <a class=‚ÄútocSkip‚Äù></a>

Es recomendable escalar los datos num√©ricos en un proyecto de ciencia de datos porque muchos algoritmos de aprendizaje autom√°tico son sensibles a la magnitud de los valores. Si las variables num√©ricas tienen escalas muy diferentes, las de mayor rango pueden dominar el aprendizaje del modelo, lo que lleva a resultados sesgados o menos precisos. El escalado asegura que todas las caracter√≠sticas contribuyan de forma equilibrada al modelo y mejora la eficiencia del entrenamiento.

</div>

```python
for depth in range(1, 11):
    model = DecisionTreeClassifier(max_depth=depth, random_state=12345)
    model.fit(X_train, y_train)
    predictions = model.predict(X_valid)
    accuracy = accuracy_score(y_valid, predictions)
    print(f'Max depth = {depth}: Accuracy = {accuracy:.3f}')
```

```python
for est in range(10, 51, 10):
    model = RandomForestClassifier(n_estimators=est, random_state=12345)
    model.fit(X_train, y_train)
    predictions = model.predict(X_valid)
    accuracy = accuracy_score(y_valid, predictions)
    print(f'Estimators = {est}: Accuracy = {accuracy:.3f}')
```

```python
# Unimos los datos de entrenamiento y validaci√≥n
X_final_train = pd.concat([X_train, X_valid])
y_final_train = pd.concat([y_train, y_valid])

# Entrenamos de nuevo (ejemplo con RandomForest)
best_model = RandomForestClassifier(n_estimators=40, random_state=12345)
best_model.fit(X_final_train, y_final_train)

# Predecimos sobre el test
test_predictions = best_model.predict(X_test)
test_accuracy = accuracy_score(y_test, test_predictions)

print(f'Final test accuracy: {test_accuracy:.3f}')
```

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(strategy='most_frequent')
dummy.fit(X_train, y_train)
dummy_predictions = dummy.predict(X_valid)
dummy_accuracy = accuracy_score(y_valid, dummy_predictions)

print(f'Dummy Classifier Accuracy: {dummy_accuracy:.3f}')
```
